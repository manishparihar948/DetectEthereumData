```python
final_result =  json_normalize(df['result'])
print(final_result)
```

```python
[df[1:]]
```

```python
df = pd.DataFrame(final_result)
```

```python
blockNumber_list = df["blockNumber"].tolist()
blockNumber_list
```

```python
timeStamp_list = df["timeStamp"].tolist()
timeStamp_list
```

```python
blockReward_list = df["blockReward"].tolist()
blockReward_list
```

```python
import csv
```

```python
from itertools import zip_longest

d = [blockNumber_list,timeStamp_list,blockReward_list]
export_data = zip_longest(*d, fillvalue = '')
with open('blockMined.csv', 'w', encoding="ISO-8859-1", newline='') as myfile:
      wr = csv.writer(myfile)
         wr.writerow(("blockNumber", "timeStamp","blockReward"))
            wr.writerows(export_data)
myfile.close()
```
