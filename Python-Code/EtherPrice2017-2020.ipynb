```python
# import library
import pandas as pd
import numpy as np
```

```python
# import data-set
data = pd.read_csv("/Users/manishparihar/Desktop/Thesis_Data/EtherPrice.csv") 
data
```

```python
data['Date(UTC)'] = pd.to_datetime(data['Date(UTC)'])
```

```python
# Filter date
start_date = '1-1-2017'
end_date = '12-31-2020'
```

```python
mask = (data['Date(UTC)'] > start_date) & (data['Date(UTC)'] <= end_date)
```

```python
data = data.loc[mask]
data
```

```python
%matplotlib inline

import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k')
#figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')



# plotting the points  
#plt.plot(data['Date(UTC)'], data['Value'], marker='o', markerfacecolor='blue', markersize=3) 
  
# plotting points as a scatter plot 
plt.scatter(data['Date(UTC)'], data['Value'], label= "stars", color= "green",  
            marker= "o", s=3) 
```

```python
# New Mask for transaction of last 15 days
start_last15days_date = '1-1-2020'
end_last15days_date = '1-31-2020'
```

```python
mask_last15days = (data['Date(UTC)'] > start_last15days_date) & (data['Date(UTC)'] <= end_last15days_date)
```

```python
data_last15days = data.loc[mask_last15days]
data_last15days
```

```python
figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k')


plt.scatter(data_last15days['Date(UTC)'], data_last15days['Value'])
plt.show()
```

# Graph of last one month

```python


import sys
df=pd.DataFrame({'x': data_last15days['Date(UTC)'], 'y': data_last15days['Value'] })
 
figure(num=None, figsize=(10, 10), dpi=100, facecolor='w', edgecolor='k')


# plot
plt.plot( 'x', 'y', data=df, marker='o', color='midnightblue',alpha=0.5)
plt.title('Ether price in last one month')
plt.xlabel('')
plt.ylabel('Ether Price per Day')

# Saving graph in png
plt.savefig('figure_LastOneMonth.png')
```
